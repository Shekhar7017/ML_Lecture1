# KNN Model 
It is works, based on neighbor's behaviour 

```python
from sklearn.neighbors import KNeighborsClassifier

X = [[180, 7],
     [200, 7.5],
     [250, 8],
     [300, 8.5],
     [330, 9],
     [360, 9.5]
    ]

# 0 = Apple, 1 = Orange
y = [0, 0, 1, 1, 1, 1]

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

weight = float(input("Enter the weight of the fruit: "))
size = float(input("Enter the size of the fruit: "))

result = model.predict([[weight, size]])[0] # with [0] we can get the first and only result from the array of predictions###

if result == 0:
    print("This is likely an Apple.")

if result == 1:
    print("This is likely an Orange.")


```

# Overfitting , Underfitting and Goodfitting


### UnderFitting
-  The model is too simple and fails to capture important patterns in the data.
#### Symptoms of underfitting
-  Poor Performance on both training &  text data.
-  High bias (model makes overly simplistic asumptions.)
#### Causes of underfitting
-  Model complexity too low (eg- used linear regression for highly non-linear data)
-  insufficient trainig or missing features.

##### Ex:- Trying to predict house price using only the numbers of rooms, ignoring location, size, etc.



### OverFitting
- The model is too complex and learns not only the true patterns but also the noise/random quirks in the training data.

#### Sypmtoms of overfitting
- Very high accuracy on training data.
- Poor generalization to test/unseen data.
- High variance (model predictions fluctuate too much).

#### Causes of overfitting
- Too many parameters/features.
- Lack of regularization.
- Small training dataset.

Ex:- A decision tree that memorizes every training sample instead of learning general rules.

### GoodFitting
- The model is just right - complex enough to capture real patterns but not so complex that it memorizes noise.

#### Symptoms of goodfitting
- Balanced performance: good accuracy  on both training and test sets.
- Low bias and low variance.

#### Goal 
- Achieve this balance by tuning hyperparameters, using cross-validation, and applying techniques like regulization or pruning.

# --------------------- Model Evalution Matrics ---------------------

Model evaluation metrics are the quantitative measures we use to judge how well a machine learning model performs. They help us understand whether the model is accurate, reliable, and suitable for real-world use. Different tasks (classification, regression, clustering) require different metrics.



##### Common Model Evalution Metrics
1. For Classification Models
    - Accuracy -> Proportion of correct predictions out of all predictions.
    - Precision -> Of the predicted positives, how many are actually positive.
    - Recall(sensitivity) -> Of the positives, How many did the model correctly identify.
    - F1 Score -> Harmonic mean of precision and recall, balances both.
    - ROC-AUC -> Measures how well the model distinguishes between classes.

2. For Regression Models
    - Mean Absolute Error(MAE) imp. -> Average absolute difference between predicted and actual values.
    - Mean Squared Error(MSE) imp. -> Average squared difference; penalizes larger errors more.
    - Root Mean Squared Error -> Square root of MSE, interpretable in original units.
    - R<sup>2</sup> (coefficient of Determination) -> Explains how much variance in the target is captured by the model.

3. For Clustring Models
    - Silhouette Score -> Measures how similar an object is to its own cluster compared to other clusters.
    - Davies-Bouldin Index -> Lower Indicate better clustering.
    - Adjusted Rand Index (ARI) -> Compares clustering results with ground truth labels.
